```python
# importing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds
```

```python
ds.info()
```

```python
#Employee count per Department 

sns.countplot(data=ds, x='Department', palette='Set2')
plt.title('Employee Count per Department')
```

finance department having more employees count followed by markating and engineering department

```python
sns.countplot(data=ds, x='Gender', palette='coolwarm')
plt.title('Gender Distribution')
```

female employee count is more than male employee

```python
# employee location distribution

sns.countplot(data=ds, x='Location', palette='Spectral')
plt.title('Number of employees by Location')
plt.xticks(rotation=45)
```

number of employee at seattle followed by new york and san francisco

```python
# age distribution

sns.histplot(ds['Age'], bins=10, kde=True)
plt.title('Age Distribution')
```

employee in 45-50 years are more followed by 25-30

```python
# year of experience Distribution

sns.histplot(ds['Years_of_Experience'], bins=10, kde=True)
plt.title('Years_of_Experience')
```

employee between 2.5 to 5 years experience are more

```python
#Monthly salary distribution

sns.boxplot(data=ds, y='Monthly_Salary', color='lightblue')
plt.title('Salary Spread')
```

20000 is minimum and 120000 is maximum salary will distributed and 75000 is average salary for distribution

```python
# average salary by department

sns.histplot(ds['Years_of_Experience'], bins=10, kde=True)
plt.title('Average salary by Department')
```

finance department has more average salary

```python
#Performance score by Department

sns.boxplot(data=ds, x='Department', y='Performance_Score', palette='pastel')
plt.title('Performance score by Department')
```

```python
# salary distribution by Department

sns.violinplot(data=ds, x='Gender', y='Monthly_Salary', palette='muted')
plt.title('Salary by Gender')
```

```python
sns.heatmap(ds[['Age', 'Years_of_Experience', 'Monthly_Salary', 'Performance_Score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
# salary scatter plot

sns.scatterplot(data=ds, x='Years_of_Experience', y='Monthly_Salary', hue='Department')
plt.title('Salary vs. Experience by Department')
```
